```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
X_train_path = 'X_train_scaled.csv'
X_test_path = 'X_test_scaled.csv'
y_train_path = 'y_train.csv'
y_test_path = 'y_test.csv'
```

```python
X_train_scaled = pd.read_csv(X_train_path)
X_test_scaled = pd.read_csv(X_test_path)
y_train = pd.read_csv(y_train_path).values.flatten()
y_test = pd.read_csv(y_test_path).values.flatten()  
```

```python
modelo = LinearRegression()
modelo.fit(X_train_scaled, y_train)
```

```python
previsoes = modelo.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test, previsoes)
mae = mean_absolute_error(y_test, previsoes)
r2 = r2_score(y_test, previsoes)
```

```python
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R² Score:", r2)
```

```python
erro_medio = mae
porcentagem_acerto = 100 - erro_medio / y_test.mean() * 100
print("Porcentagem de Acerto:", porcentagem_acerto)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, previsoes, alpha=0.5, edgecolors='w', s=80)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Valores Reais vs. Previsões')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.grid(True)
plt.show()
```

```python
erros = y_test - previsoes

plt.figure(figsize=(10, 6))
sns.histplot(erros, bins=30, kde=True)
plt.xlabel('Erro')
plt.title('Distribuição dos Erros')
plt.grid(True)
plt.show()
```
